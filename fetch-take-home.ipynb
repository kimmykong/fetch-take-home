*Specifications*

Provide routes that:
 - Add transactions for a specific payer and date
   - POST (payer,date) -> "Request processed"
 - Spend points using the rules above and return a list of {"payer":<string>,"points":<integer>}
   - POST (points) -> [{"payer":payerName,"points": - spentPoints}]
 - Return all payer point balances
   - GET (points) -> {"payerName":points}


```python
kim_IP = "127.0.0.1:5000"
```

```python
import requests

address = "http://"+kim_IP
# Add transactions from example, call get balances, print response of the call
requests.post(address+"/addTransaction", data={"name": "DANNON", "points":1000, "timestamp":"2020-11-02T14:00:00Z"})
requests.post(address+"/addTransaction", data={"name": "UNILEVER", "points":200, "timestamp":"2020-10-31T11:00:00Z"})
requests.post(address+"/addTransaction", data={"name": "DANNON", "points":-200, "timestamp":"2020-10-31T15:00:00Z"})
requests.post(address+"/addTransaction", data={"name": "MILLER COORS", "points":10000, "timestamp":"2020-11-01T14:00:00Z"})
requests.post(address+"/addTransaction", data={"name": "DANNON", "points":300, "timestamp":"2020-10-31T10:00:00Z"})
balance = requests.get(address+"/getPayerBalances")
print ("Add transactions from example, call get balances, print response of the call")
print("Balances:",balance.text)

# Spend 5000 points and print response of the call
spend5000 = requests.post(address+"/spendPoints", data={"points": 5000})
print("Spend 5000 points and print response of the call")
print(spend5000.text)

# Print response of get balances
balance = requests.get(address+"/getPayerBalances")
print("Print response of get balances")
print("Balances:",balance.text)

# Spend more points than exist in the system and print response of the spend
spend50000 = requests.post(address+"/spendPoints", data={"points": 50000})
print("Spend more points than exist in the system and print response of the spend")
print(spend50000.text)

# Print response of get balances
balance = requests.get(address+"/getPayerBalances")
print("Print response of get balances")
print("Balances:",balance.text)

# Try to spend when the balances are zero
spend500 = requests.post(address+"/spendPoints", data={"points": 500})
print("Try to spend when the balances are zero, print response of overspend, print balances")
print(spend500.text)
balance = requests.get(address+"/getPayerBalances")
print("Balances:",balance.text)
```
